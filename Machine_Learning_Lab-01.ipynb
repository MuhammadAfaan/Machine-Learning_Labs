Task 3

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="sex", hue="survived")
plt.title("Survival Rate by Gender(M/F)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="class", hue="survived")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], kde=True, bins=30)
plt.axvline(df['age'].median(), color='red', linestyle='dashed', label="Median Age")
plt.legend()
plt.title("Age Distribution using Mediun")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="class", y="fare")
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
numeric_data = df.select_dtypes(include=['number'])
corr_matrix = numeric_data.corr()


plt.figure(figsize=(8,5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Data Cleaning Steps

To prepare the dataset for analysis, the following steps were performed:

### Handling Missing Values:
- Missing values in the `age` column were filled with the median age.
- Rows with missing values in the `embark_town` column were removed.

### Feature Engineering:
- A new `family_size` feature was created by combining `sibsp` and `parch` columns.
- A `travel_alone` feature was added to indicate if a passenger traveled alone.

### Data Filtering:
- A subset of passengers aged between 18 and 50 was created to focus on a specific age group.

---

## Insights from EDA and Visualizations

### Survival Rate by Gender:
- Women had a higher survival rate than men.
- First-class passengers had the highest survival rates, while third-class passengers had the lowest.

### Age Distribution:
- The median age of passengers was identified.

### Fare Analysis:
- First-class passengers paid significantly higher fares than those in lower classes.

## Observations on Patterns and Anomalies

### Gender-Based Survival Bias:
- Women survived at a higher rate, following the "women and children first" protocol.

### Class Disparities:
- First-class passengers had better survival outcomes, highlighting social inequality.

### Outliers in Fare Prices:
- Some extreme fare values were seen, likely from wealthy passengers or premium accommodations.

### Travel Alone vs. With Family:
- Passengers traveling alone had a lower survival rate, suggesting group travel offered some advantage.

### NYC Airbnb Listings:
- Listings were mostly concentrated in Manhattan and Brooklyn, popular tourist areas.
- Price distribution showed significant outliers, possibly from luxury accommodations or pricing errors.
